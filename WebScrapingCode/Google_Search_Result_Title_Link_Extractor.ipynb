```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

#ドライバーの立ち上げ
driver = webdriver.Firefox()

query_list = ['ラーメン二郎','伝説のすた丼','春日亭','蒙古タンメン中本','吉野家','日高屋']

#Googleのトップ画面を開く。
driver.get('https://www.google.co.jp/')

#取得したデータを格納する箱を用意
query_data = []
title_data  = []
link_data  = []

for query in query_list:
    #検索語を入力して送信する。
    input_element = driver.find_element_by_name('q')
    #検索窓で文字をクリアする
    input_element.clear()
    #キーワードを入力する。
    input_element.send_keys(query)
    #エンターキーを押す
    input_element.send_keys(Keys.RETURN)
    
    time.sleep(10)
    
    #検索結果からデータを抽出する、
    for a in driver.find_elements_by_css_selector('div.rc'):
        for link in a.find_elements_by_css_selector('h3 > a'):
            time.sleep(1)
            #データを変数に追加していく。
            query_data.append(query)
            title_data.append(link.text)
            link_data.append(link.get_attribute('href'))
            serp_data = zip(query_data,title_data,link_data)
            dataset = pd.DataFrame(list(serp_data))
            dataset.columns = ['query','title','url']
            #データをCSVで出力する。
            dataset.to_csv('serp_result.csv')

    time.sleep(30) #60秒空けたら安心できると思う。
```
